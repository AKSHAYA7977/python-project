```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("Downloads/customers.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
#no null values present
```

```python
plt.scatter(data["Age"],data["Spending Score (1-100)"])
plt.xlabel("AGE")
plt.ylabel("Spending Score")
plt.title("AGE VS SPENDING SCORE")
plt.show()
```

```python

X = data[['Age' , 'Spending Score (1-100)']].values
ar = []
for n in range(1 , 11):
    model= KMeans(n_clusters = n)
    model.fit(X)
    ar.append(model.inertia_)
```

```python
pd.DataFrame(ar).plot()
plt.grid()
```

```python
#so the maximum no of clusters should be 3
model = KMeans(n_clusters = 3)
model.fit(X)
labels = model.labels_
```

```python
labels
```

```python
clusters = data[['Age', 'Spending Score (1-100)']]
```

```python
clusters = clusters.assign(cluster=labels)
```

```python
clusters.plot.scatter(x='Age', y='Spending Score (1-100)', c='cluster', colormap='viridis')
```
